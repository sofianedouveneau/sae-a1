### Question 2

Définissez la fonction `eleves` prenant en paramètre un tableau de réponses et retournant un tableau contenant les élèves (chaînes de caractères) ayant répondu aux questionnaire. L'appel de cette fonction avec le tableau de réponses donné en exemple doit retourner le tableau :


Définissez une fonction de tests unitaires pour la fonction `eleves`.

### Question 3

Définissez la fonction `lecture_reponses` prenant en paramètre un nom de fichier et retournant un tableau de réponses en fonction des données contenues dans le fichier. Ce dernier comporte une ligne par élève. Chaque ligne a le format `nom:R1/R2/R3/R4` où `nom` correspond au nom de l'élève et `R1`, `R2`, `R3`, `R4` correspondent aux scores (réponses aux questions) de l'élève. Par exemple, l'appel de la fonction `lecture_reponses` avec un nom de fichier contenant 
```
Harry Potter:10/5/8/1
Cedric Diggory:6/7/9/4
Drago Malefoy:1/3/2/10
```
doit retourner le tableau de réponses donné en exemple.

Définissez une fonction de tests unitaires pour la fonction `lecture_reponses`.

````python
############################## 
#   Saisir votre code ici.   #
##############################

path = '/home/sosocrates/Desktop/R101_1/R101_1/SAE_101/questionnaire_premiere_annee.txt'
mode = 'r'

f = open(path, mode) # ouverture du fichier afin de pouvoir se servir du questionnaire_premiere_annee.txt
#tab = f.readlines() : tab originel
tab = ["Harry Potter:10/5/8/1","Cedric Diggory:6/7/9/4","Drago Malefoy:1/3/2/10"] # tableau utilisé pour les asserts
reponses = ["Harry Potter", 10, 5, 8, 1, "Cedric Diggory", 6, 7, 9, 4, "Drago Malefoy", 1, 3, 2, 10] # résultat à obtenir


#print(tab)

def lecture_reponses(tab) :
    
    """ 
        Lis les réponses au questionnaire et sépare les résultats des prenoms pour les attribuer 
        à différentes indices du tableau.
        Résultat cherché : ['Harry Potter', 10, 5, 8, 1, 'Cedric Diggory', 6, 7, 9, 4, 'Drago Malefoy', 1, 3, 2, 10]
    """
    
    i = 0
    tab_reponses = []
    
    while i!=len(tab):
        tab_strip = tab[i].strip() # enlève les \n aux indices [i] du questionnaire
        if ':' in tab_strip : # s'il y a ":" de présent dans la variable tab_strip
            
            tab_split = tab_strip.split(':') # enlève le séparateur ":"
            nom = tab_split[0] # chaque i=0 correspond aux noms d'élèves, ça sera attribué à la variable nom
            tab_reponses.append(nom) # nom est ajouté au tableau
            
            reponses = tab_split[1] # chaque i= 1 correspond aux réponses, ça sera attribué à la variable reponses
            rep_split = reponses.split('/') # on supprime le séparateur "/" de la chaine de charactère correspondant aux réponses
            
            ib = 0
            while ib < len(rep_split):
                tab_reponses.append(int(rep_split[ib])) # convertit les réponses en entiers pour ensuite les ajouter au tableau
                ib += 1
        i+=1
    return tab_reponses

def test_lecture_reponses():
    
    """ 
        Confirme le bon fonctionnement de la fonction
    """ 
    
    assert lecture_reponses([]) == [], "tableau vide" # si le tableau est vide, le résultat recherché devrait être []
    assert lecture_reponses(tab) == reponses # si le tableau est plein, le résultat devrait correspondre à la variable reponses
    print("Fonction marche")

test_lecture_reponses()
print(lecture_reponses(tab))

# ?lecture_reponses : affiche le docstring de la fonction lecture_reponses
````

### Question 4

Définissez la fonction `maison` prenant en paramètre un tableau de réponses ainsi que l'indice d'un élève. La fonction retourne la maison correspondant à l'affectation de cet élève. Ainsi, l'appel de la fonction maison avec le tableau de réponses donné en exemple doit retourner : 
- `"Gryffondor"` avec l'indice `0` (correspondant à Harry Potter),
- `"Poufsouffle"` avec l'indice `5` (correspondant à Cedric Diggory),
- `"Serpentard"` avec l'indice `10` (correspondant à Drago Malefoy).

En effet, Harry Potter a mis le meilleur score à la question 1 et doit donc être affecté à Gryffondor. Cedric Diggory a mis le meilleur score à la question 3 et doit être affecté à la maison Poufsouffle. Finalement, Drago a mis le meilleur score à la question 4 et doit donc être affecté à la maison Serpentard.

Définissez également une fonction de tests unitaires pour la fonction `maison`.

````python
############################## 
#   Saisir votre code ici.   #
##############################

tab_reponses = lecture_reponses(tab)
indice_eleve = 10 # indice à Harry Potter

def maison(tab_reponses, indice_eleve):
    
    """
        Détermine la maisons Poudlard de chaque élèves.
        Les quatres questions correspondant respectivement à:
        Gryffondor, Serdaigle, Poufsouffle et Serpentard
        
        On cherche à comparer les valeurs associées aux réponses pour renvoyer la maison du score
        le plus élevé.
    """
    
    # on récupère les réponses aux questions
    q1 = tab_reponses[indice_eleve + 1]
    q2 = tab_reponses[indice_eleve + 2]
    q3 = tab_reponses[indice_eleve + 3]
    q4 = tab_reponses[indice_eleve + 4]
    
    # on compare pour retrouver la valeur la plus élevée et on détermine la maison
    if q1>=q2 and q1>=q3 and q1>=q4:
        return "Gryffondor" 
    if q2>=q1 and q2>=q3 and q2>=q4:
        return "Serdaigle"
    elif q3>=q1 and q3>=q2 and q3>=q4:
        return "Poufsouffle"
    else: 
        return "Serpentard"


print(maison(tab_reponses, indice_eleve))

tab = ['Harry Potter', 10, 5, 8, 1, 'Cedric Diggory', 6, 7, 9, 4, 'Drago Malefoy', 1, 3, 2, 10]

def test_maison():
    
    """ 
        Confirme le bon fonctionnement de la fonction maison
    """
    
    assert maison(tab, 0) == "Gryffondor" # la maison à Potter attendue est Gryffondor
    assert maison(tab, 5) == "Poufsouffle" # la maison à Diggory attendue est Poufsouffle
    assert maison(tab, 10) == "Serpentard" # la maison à Malfoy attendue est Serpentard
    print("tout est ok")
    
test_maison()

#?maison : affiche de docstring de la fonction maison
````

### Question 5

Définissez la fonction `repartition` prenant en paramètre un tableau de réponses et retournant un dictionnaire dont les clés sont les noms des élèves et les valeurs la maison à laquelle les élèves sont affectés. Par exemple, l'appel de la fonction `repartition` avec le tableau de réponses donné en exemple doit retourner le dictionnaire : 
```python
{
    "Harry Potter": "Gryffondor",
    "Cedric Diggory": "Poufsouffle",
    "Drago Malefoy": "Serpentard"
}
```

Définissez également une fonction de tests unitaires pour la fonction `repartition`.

````python
############################## 
#   Saisir votre code ici.   #
##############################

def repartition(tab_reponses):
    
    """
        Répartit les élèves dans leur maisons respectivesen fonction de leurs réponse.
        
        On associe l'indice de l'élève avec sa maison obtenue dans la fonction maison ci-dessus
    """
    
    
    indice_eleve = 0 # initialise le compteur pour pourcourir les réponses des dicos
    
    dico_maisons = {} # dictionnaire final demandé
    
    while indice_eleve!=len(tab_reponses):
        
        dico_maisons[tab_reponses[indice_eleve]] = maison(tab_reponses, indice_eleve) # ajoute la maison à l'indice correspondante
        indice_eleve += 5  # on passe à l'élève prochain
    
    return dico_maisons # retourne le résultats

print(repartition(tab_reponses))

def test_repartition():
    
    """ 
        Confirme le bon fonctionnement de la fonction repartition
    """
    
    assert repartition(tab_reponses) == {'Harry Potter': 'Gryffondor', 'Cedric Diggory': 'Poufsouffle', 'Drago Malefoy': 'Serpentard'}
    print("La fonction marche")
    
test_repartition()

# ?repartition : affiche le docstring de la fonction repartition
````

Définir la fonction `nb_erreurs` prenant en paramètre deux dictionnaires dont les clés sont des élèves (les mêmes dans les deux dictionnaires) et les valeurs les maisons auxquelles ils sont affectés. La fonction renvoie le nombre "d'erreurs", c'est-à-dire le nombre de fois où les valeurs des deux dictionnaires sont différentes pour un même élève. Par exemple, l'appel de la fonction `nb_erreurs` avec les deux dictionnaires
```python
{"Harry Potter": "Gryffondor", "Cedric Diggory": "Poufsouffle", "Drago Malefoy": "Serpentard"}

{"Cedric Diggory": "Serdaigle", "Drago Malefoy": "Gryffondor", "Harry Potter": "Gryffondor"}
```
doit retourner 2.


Définir une fonction de tests unitaires pour la fonction `nb_erreurs`.


À l'aide des deux fichiers de données, afficher le pourcentage d'erreur de votre méthode, c'est-à-dire le nombre d'élèves de première année que vous avez affectés à une maison différente que celle choisie par le choixpeau magique.

````python
############################## 
#   Saisir votre code ici.   #
##############################

import json

def nb_erreurs(d1, d2):
    
    """
        Compare les deux dictionnaires d'affectation et renvoie le nombre d'erreurs:
        cad le nombre de fois où les maisons diffèrent pour un même élève.
    """
    
    erreurs = 0  # compteur d'erreurs
    cles = list(d1.keys())  # on récupère toutes les clés (noms des élèves)
    
    i = 0
    while i < len(cles):
        nom = cles[i]
        if d1[nom] != d2[nom]:  # si les maisons sont différentes
            erreurs += 1 
        i += 1

    return erreurs

def test_nb_erreurs():
    
    """
        S'assure le bon fonctionnement de la fonction nb_erreurs
    """
    
    d1 = {"Harry Potter": "Gryffondor", "Cedric Diggory": "Poufsouffle", "Drago Malefoy": "Serpentard"}
    d2 = {"Cedric Diggory": "Serdaigle", "Drago Malefoy": "Gryffondor", "Harry Potter": "Gryffondor"}
    assert nb_erreurs(d1, d2) == 2, "devrait renvoyer 2 erreurs"

    print("Tot est ok")

def calcul_pourcentage_erreur():
    
    """
        Calcule le pourcentage d'erreurs entre les maisons attribuées
        par la fonction `repartition` et celles enregistrées dans le fichier JSON.
    """
    
    # lecture
    f = open("questionnaire_premiere_annee.txt", "r")
    tab = f.readlines()
    f.close()

    # transforme les réponses en tableaux
    tab_reponses = lecture_reponses(tab)
    dico_maison = repartition(tab_reponses)

    # lecture du fichier json
    f = open("affectation_premiere_annee.json", "r")
    dico_choixpeau = json.load(f)
    f.close()

    # calcul des erreurs
    erreurs = nb_erreurs(dico_maison, dico_choixpeau)
    total = len(dico_maison)
    pourcentage = erreurs / total * 100
    
    print(nb_erreurs(dico_maison, dico_choixpeau))
    print("Nombre d'élèves :", total)
    print("Erreurs :", erreurs)
    print("Pourcentage d'erreur :", pourcentage, "%")


calcul_pourcentage_erreur()

````

### Question 7

Suite aux ricanements de Severus, le professeur Remus Lupin vous propose de tester la fiabilité d'une méthode de répartition aléatoire des élèves. Pour cela, répartissez aléatoirement les élèves dans les maisons et comptabiliser le nombre d'erreurs avec la répartition du choixpeau. Vous ferez la moyenne sur 100 répartitions aléatoires pour avoir une valeur plus précise. Que peut-on en conclure ?

````python
############################## 
#   Saisir votre code ici.   #
##############################

from json import *
from random import *

def repartition_aleatoire(tab):
    """
        Répartit aléatoirement les élèves dans les maisons de Poudlard.
    """
    
    maisons = ["Gryffondor", "Serdaigle", "Poufsouffle", "Serpentard"]
    dico = {}
    i = 0

    while i < len(tab):
        ligne = tab[i].strip()  # enlève les espaces et retours à la ligne
        if ":" in ligne:        # vérifie que la ligne contient bien un nom
            nom = ligne.split(":")[0] # enlève ':' devantle nom
            dico[nom] = choice(maisons)  # attribution aléatoire d'une maison
        i += 1
    
    return dico
    
path = '/home/sosocrates/Desktop/R101_1/R101_1/SAE_101/questionnaire_premiere_annee.txt'
mode = 'r'

f = open(path, mode) # ouvre le fichier questionnaire_premiere_annee.txt
tab = f.readlines() # convertit le fichier en tableau
f.close()

f = open("affectation_premiere_annee.json", "r")
dico_choixpeau = json.load(f) # ouvre le fichier json
f.close()

total_eleves = len(dico_choixpeau)
total_pourcentage = 0

essais = 0 # initialise le nb d'essais
while essais < 100:
    
        # répartition aléatoire d'élèves dans les maisons
        dico_random = repartition_aleatoire(tab)
        
        # calcul du nombre d'erreurs entre la répartition aléatoire et celle du choixpeau
        erreurs = nb_erreurs(dico_random, dico_choixpeau)
        
        # calcul du pourcentage d'erreurs
        pourcentage = erreurs / total_eleves * 100
        
        # ajout du pourcentage au total cumulé
        total_pourcentage += pourcentage
        
        # incrémentation des essais
        essais += 1


moyenne_pourcentage = total_pourcentage / 100 # calcule la moyenne


print("Nombre d'élèves :", total_eleves)
print("Pourcentage d'erreur moyen sur 100 répartitions aléatoires :", moyenne_pourcentage, "%")


if moyenne_pourcentage <  60:
    print("La méthode de répartition aléatoire est très peu fiable.")
else:
    print("La méthode aléatoire est proche de celle du choixpeau.")

````
